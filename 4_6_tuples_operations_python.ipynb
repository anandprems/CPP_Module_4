#Tuple - list

```python
my_tuple9 = (4,7,3,['rocky','yashin','malik'])

print(my_tuple9)

# access list present inside tuple
print(my_tuple9[3])

# delete last eleemnt of list
my_tuple9[3].pop()
print(my_tuple9)

# add in list
my_tuple9[3].append('roja')

print(my_tuple9)

# clear list
my_tuple9[3].clear()

print(my_tuple9)
```

#Tuple - Zip function

```python
fruits = ['mango', 'orange', 'grapes']
colors = ['yellowish-orange','orange','black']

result = zip(fruits, colors)
print(list(result))
```

```python
alpha_ = ('a','b','c','d')
animal_ = ['lion','tiger','monkey','zebra']
bird_ = ['pigeon','eagle','peacock']
```

```python
result = zip(alpha_, animal_)
result_list = list(result)
print(result_list)
```

```python
result1 = zip(bird_, animal_)
result_set = set(result1)
print(result_set)
```

```python
result2 = zip(alpha_, bird_)
result_list = set(result2)
print(result_list)
```

```python
#unzipping the value of zipped function 
alpha, bird = zip(*result_list)
print(alpha)
print(bird)
```
